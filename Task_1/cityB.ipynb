```python
import numpy as np 
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
poi_data = pd.read_csv('../Datasets/Task 1/POIdata_cityB.csv')
categories = pd.read_csv('../Datasets/Task 1/POI_datacategories.csv', header=None)
```

```python
poi_data['grid_id'] = poi_data['x'].astype(str) + "_" + poi_data['y'].astype(str)
grid_baskets = poi_data.groupby('grid_id')['category'].apply(list)

for basket in grid_baskets:
    basket = list(set(basket))

grid_baskets
```

```python
one_hot_list = []

for i in range(len(grid_baskets)):
    row = []
    for j in range(len(categories)):
        row.append(0)
    one_hot_list.append(row)


for basket_index in range(len(grid_baskets)):
    for category_index in grid_baskets[basket_index]:
        one_hot_list[basket_index][category_index-1] = 1

one_hot_df = pd.DataFrame(one_hot_list, columns=categories[0])
one_hot_df
```

```python
grid_baskets_df = grid_baskets.reset_index()
grid_baskets_df.columns = ['grid_id', 'categories']
print(grid_baskets_df)
```

```python
grid_baskets_df = grid_baskets_df.join(one_hot_df)
grid_baskets_df.drop(['categories'], axis='columns', inplace=True)
grid_baskets_df.set_index('grid_id', inplace=True)
grid_baskets_df
```

```python
freq_items = apriori(grid_baskets_df, min_support=0.1, use_colnames=True)

rules = association_rules(
    freq_items, metric="lift", min_threshold=1, num_itemsets=len(grid_baskets)
)
rules = rules.sort_values(["confidence", "lift"], ascending=[False, False])
print(rules.head())
```

### Top frequent itemsets
and verification of support values

```python
freq_itemsets_copy = freq_items.copy()

# Calculate the support counts manually and add as a new column
support_counts = []
for index, row in freq_itemsets_copy.iterrows():
    itemset = list(row["itemsets"])  # convert frozenset of items to list
    items = one_hot_df[itemset]  # get the columns in the itemset
    mask = items.all(
        axis=1
    )  # check if all items in the itemset are present in each row (grid cell)
    support_count = mask.sum() / len(one_hot_df)
    support_counts.append(support_count)

freq_itemsets_copy["check_support"] = support_counts
# sort the dataframe by the support count and reset the index

freq_itemsets_copy = freq_itemsets_copy.sort_values(
    "support", ascending=False
).reset_index(drop=True)

freq_itemsets_copy.head()
```

```python
# save to csv
freq_itemsets_copy.to_csv("../Datasets/Task 1/freq_itemsets_cityB.csv")
```

### Top Association Rules by Lift

```python
# keep only antecedents, consequents and lift
rulesB = rules.copy()[["antecedents", "consequents", "lift"]]
# combine antecedents and consequents into a rule column with this format: (antecedents) -> (consequents)
rule_col = (
    rulesB["antecedents"].apply(lambda x: ", ".join(list(x)))
    + " -> "
    + rulesB["consequents"].apply(lambda x: ", ".join(list(x)))
)
# insert rule column into df
rulesB.insert(0, "rule", rule_col)
# drop antecedents and consequents
rulesB.drop(["antecedents", "consequents"], axis="columns", inplace=True)
# sort by lift and reset index
rulesB = rulesB.sort_values("lift", ascending=False).reset_index(drop=True)
```

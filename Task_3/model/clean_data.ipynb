```python
import os
import pandas as pd
```

```python
base_path = "../../Datasets/Task 2/"
paths = {"A": os.path.join(base_path, "frequent_sequences_A.csv"),
         "B": os.path.join(base_path, "frequent_sequences_B.csv"),
         "C": os.path.join(base_path, "frequent_sequences_C.csv"),
         "D": os.path.join(base_path, "frequent_sequences_D.csv")}
```

```python
def load_data(city):
    data = pd.read_csv(paths[city])
    return data
```

```python
def remove_dup_triplegs(data):
    # Make a copy of the data to avoid modifying the original dataframe
    data = data.copy()

    # Convert the 'Pattern' column from string representation of tuples to actual tuples
    data["Pattern"] = data["Pattern"].apply(eval)

    # Define a function to remove consecutive duplicate tuples in a list of tuples
    def remove_consecutive_duplicates(pattern):
        result = []
        for i in range(len(pattern)):
            if i == 0 or pattern[i] != pattern[i - 1]:
                result.append(pattern[i])
        return tuple(result)

    # Apply the function to the 'Pattern' column
    data["Pattern"] = data["Pattern"].apply(remove_consecutive_duplicates)

    # find indices of rows with list of length 1
    indices = data[data["Pattern"].apply(len) == 1].index
    data.drop(indices, inplace=True)

    return data
```

```python
for city in ["A", "B", "C", "D"]:
    df = load_data(city)
    display(df.head())
    print(f"{city} shape before: {df.shape}")
    df_clean = remove_dup_triplegs(df)
    print(f"{city} shape after: {df_clean.shape}")
    display(df_clean)
    # save cleaned data
    df_clean.to_csv(f"{paths[city].replace('.csv', '_cleaned.csv')}", index=False)
```
